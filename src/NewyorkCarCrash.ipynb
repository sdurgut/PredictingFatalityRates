```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# skim the dataset, it is quite large right now
def readCrashData():
    df = pd.read_csv("Motor_Vehicle_Crashes_-_Vehicle_Information__Three_Year_Window.csv")
    df.drop(["Partial VIN","State of Registration","Fuel Type","Type / Axles of Truck or Bus","Registration Class", "Case Vehicle ID"],inplace=True,axis=1)
#     df.drop(["ID","Case Number","Block","IUCR","FBI Code", "Updated On","Location"],inplace=True,axis=1)
#     df.to_csv("SkimmedChicagoCrime.csv")
#     print ("Data is skimmed and saved as SkimmedChicagoCrime.csv")
    return df


df = readCrashData()
```

```python
df["Year"].hist(grid=False, xrot = 45 , bins=5)
x = [2012,2013,2014]
plt.xticks(np.arange(min(x), max(x)+1, 1.0))
#plt.xlabel("Year")
plt.ylabel("Car Crashes/Year")

plt.show()
```

```python
tmp = df["Action Prior to Accident"].groupby(df["Action Prior to Accident"]).count()
fig=plt.figure(figsize=(10, 6), dpi= 80, facecolor='w', edgecolor='k')
tmp.plot.bar()
#plt.xticks(rotation="Vertical")
plt.show()

```

```python
tmp1 = df["Vehicle Make"].groupby(df["Vehicle Make"]).count()
#fig1=plt.figure(figsize=(15, 12), dpi= 80, facecolor='w', edgecolor='k')
# tmp.plot.bar()
# plt.show()
tmpdf = tmp1.to_frame()
tmpdf.columns = ["Collision"]
tmpdf = tmpdf[tmpdf["Collision"]>5000]
tmpdf.reset_index(inplace=True)
tmpdf.plot(x='Vehicle Make', y='Collision', kind='bar', legend=False)
plt.show()
```

```python

tmp2 = df["Contributing Factor 1 Description"].groupby(df["Contributing Factor 1 Description"]).count()
tmp2df = tmp2.to_frame()
tmp2df.columns = ["Count"]
tmp2df = tmp2df[tmp2df["Count"]>1000]
tmp2df.reset_index(inplace=True)
#tmp2df = tmp2df[(tmp2df["Contributing Factor 1 Description"] != "Not Entered")]
tmp2df = tmp2df[(tmp2df["Contributing Factor 1 Description"] != "Not Entered") & (tmp2df["Contributing Factor 1 Description"] != 'Not Applicable') & (tmp2df["Contributing Factor 1 Description"] != 'Unknown')]
tmp2df.reset_index(inplace=True)
#fig2=plt.figure(figsize=(15, 12), dpi= 80, facecolor='w', edgecolor='k')
tmp2df.plot(x='Contributing Factor 1 Description', y='Count', kind='bar', legend=False)
plt.show()
# tmp2df

```
